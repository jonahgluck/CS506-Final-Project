# LUFlow PCA Visualization

```python
import numpy as np
import pandas as pd 
import plotly.express as px
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
```

```python
df1 = pd.read_csv('/kaggle/input/luflow-network-intrusion-detection-data-set/2022/06/2022.06.12/2022.06.12.csv')
df1["date"] = "2022.06.12"
df2 = pd.read_csv('/kaggle/input/luflow-network-intrusion-detection-data-set/2022/06/2022.06.13/2022.06.13.csv')
df2["date"] = "2022.06.13"
df3 = pd.read_csv('/kaggle/input/luflow-network-intrusion-detection-data-set/2022/06/2022.06.14/2022.06.14.csv')
df3["date"] = "2022.06.14"



df = pd.concat([df1, df2, df3], ignore_index=True)

num_samples = 1_000

df = df.groupby("label").sample(num_samples)

df.head()
```

## Plotting Singular Values

```python
X = df[["bytes_in", "bytes_out", "avg_ipt", "entropy", "num_pkts_out", "num_pkts_in", "duration"]]

pca = PCA()

pca.fit_transform(X)

singular_values = pca.singular_values_

plt.plot([1, 2, 3, 4 ,5, 6, 7], singular_values, "ro")

plt.title("Singular Values From Most to Lease Significant")
plt.ylabel("Singular Value")

plt.show()

```

```python
num_samples = 1_000

df = df.groupby("label").sample(num_samples)
```

## PCA Plots with all features

```python
X = df[["bytes_in", "bytes_out", "avg_ipt", "entropy", "num_pkts_out", "duration"]]

pca = PCA(n_components=2)

components = pca.fit_transform(X)

total_var = pca.explained_variance_ratio_.sum() * 100

fig = px.scatter(components, x=0, y=1, color=df["label"], title=f'Total Explained Variance: {total_var:.2f}%')
fig.show(renderer="iframe")

print(f"Total Explained Variance for First PCA Component: {pca.explained_variance_ratio_[0]}")
print(f"Total Explained Variance for Second PCA Component: {pca.explained_variance_ratio_[1]}")
```

```python
pca = PCA(n_components=3)
components = pca.fit_transform(X)

total_var = pca.explained_variance_ratio_.sum() * 100

fig = px.scatter_3d(
    components, x=0, y=1, z=2, color=df["label"],
    title=f'Total Explained Variance: {total_var:.2f}%',
    labels={'0': 'PC 1', '1': 'PC 2', '2': 'PC 3'}
)
fig.show(renderer="iframe")
```
